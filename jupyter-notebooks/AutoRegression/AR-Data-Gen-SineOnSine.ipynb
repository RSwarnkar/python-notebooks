# Data Generator : Sine wave on Sine wave

```python
import numpy as np
import pandas as pd
import matplotlib
%matplotlib inline
```

```python
# x = a * sin (omega * t + w)
# omega = 2*pi/T
# T = time period of sine wave, T=2 periods

a1 = 1000
T1 = 50
omega1 = 2 * np.pi / T1
w1 = np.pi / 2
t1 = np.arange(0,1096)
fact1 = omega1 * t1 + w1

x1 = a1 * np.sin(fact1)

x1 = x1 + 100
```

```python
# x = a * sin (omega * t + w)
# omega = 2*pi/T
# T = time period of sine wave, T=2 periods

a2 = 1000
T2 = 100
omega2 = 2 * np.pi / T2
w2 = np.pi / 2
t2 = np.arange(0,1096)
fact2 = omega2 * t2 + w2

x2 = a2 * np.sin(20*fact2)

x2 = x2 + 100
```

```python
y = x1 + x2
```

```python
df=pd.DataFrame(data=y)
df.plot(figsize=(20,8))
```

```python
# y = m * x + c

m = 120
x = np.arange(0,1096)
c = np.random.randint(100,150,1096)

z = m * x + c

z
```

```python
w = y + z
```

```python
df=pd.DataFrame(data=w)
df.plot(figsize=(20,8))
```

```python
# Timeframe

timestamps = pd.date_range('2016-01-01', periods=1096, freq='D')
timestamps

df=pd.DataFrame(data=w,index=timestamps)
df.index.name = 'day'
df.columns=['sales']
df.head()
```

```python
df['2017-01-01':'2017-02-10'].plot(figsize=(12,6))
```
