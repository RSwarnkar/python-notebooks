```python
import numpy as np
import pandas as pd
import matplotlib
%matplotlib inline
```

```python
# x = a * sin (omega * t + w)
# omega = 2*pi/T
# T = time period of sine wave, T=2 periods

a = 1000
T = 50
omega = 2 * np.pi / T
w = np.pi / 2
t = np.arange(0,1096)
fact = omega * t + w

x = a * np.sin(fact)

x = x + 100

len(x)


```

```python
# Timeframe

timestamps = pd.date_range('2016-01-01', periods=1096, freq='D')
timestamps
```

```python

df=pd.DataFrame(data=x,index=timestamps)
df.index.name = 'day'
df.columns=['sales']
df.head()


```

```python
df.plot(figsize=(12,6))
```

```python
# y = m * x + c

m = 120
x = np.arange(0,1096)
c = np.random.randint(1000,1500,1096)

y = m * x + c

y
```

```python
z = x + y
z
```

```python
df2=pd.DataFrame(data=z,index=timestamps)
df2.index.name = 'day'
df2.columns=['sales']
df2['sales'] = df2['sales'] * 0.01
df2.tail()
```

```python
df2.plot(figsize=(12,6))
```
