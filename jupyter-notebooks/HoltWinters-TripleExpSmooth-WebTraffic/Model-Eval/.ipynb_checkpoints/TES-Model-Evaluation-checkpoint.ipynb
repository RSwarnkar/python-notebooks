# Timecard Traffic (Web Request Count) Prediction
### Using Holt-Winters (Triple Exponential Smoothing - TES) - Python

It was found that the RMSE of the TES in Timecard Web Traffic was much larger than the standard deviation. 
Which indicates that this is NOT A GOOD prediction result can be achieved.

Much smaller RMSE ensures nice fit on the data. However does not ensures Overfitting. 

Anyways, lets do some post predictive analysis by comparing actual data vs predicted data of TES by suing RMSEs.

```python
import numpy as np
import pandas as pd
import matplotlib
%matplotlib inline
```

```python
#read the CSV 
df = pd.read_csv('predicted-vs-real-timecard.csv', sep=';', index_col='date', parse_dates=True)
```

```python
df.head()
```

```python
# Set Frequency:
df = df.asfreq('D')
df.index
df['y_hat'].plot(figsize=(12,5),title='TES vs Actual',legend=True, label='TES Predicted (y_hat)')

df['actual'].plot(legend=True,label='Actual',figsize=(12,5))
```

# Conclusion: 
This is <span style='color:red'>UNACCEPTABLE PREDICTION</span>. and the TES fails to predict thing within the acceptable limits (limit upto order of +/-1000). 
However, the TES does <span style='color:green'>predicts quite well on the WEEKENDS</span> i.e. on Saturdays and Sundays.

Hence we need to improve upon the predictions by:
1. Identify the root cause of large RMSE than Standard Deviation
2. Check for the Exogenous variables in the Time dimension (Like holidays, Month ends)
3. Apply RNN and see if this prerforms better
