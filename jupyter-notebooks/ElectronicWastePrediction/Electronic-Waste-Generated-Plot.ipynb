# Electronic Waste Generated & Recycled Prediction
* Sources: https://catalog.data.gov/dataset?q=waste

```python
import pandas as pd
import matplotlib
%matplotlib inline
```

```python
waste = pd.read_csv("ElectronicWasteYearly.csv", index_col='Year',parse_dates=True)
```

```python
waste.head()
```

```python
waste['Recycled'].plot(figsize=(12,5))
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
result = seasonal_decompose(waste['Generated'],model='mul') # model='mul'
```

```python
result.trend
```

```python
result.seasonal
```

```python
result.resid #if there is no resid then in multiplicative it will be 1, and in additive it will be 0
```

```python
result.plot();
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 12,5
result.plot();
```

## Conclusion: There is no Seasonality in the data found here. Note: This does not mean that there should not be seasonal data with waste generation but may arise due to sparse data available.
